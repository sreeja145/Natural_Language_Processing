```python
# lowercasing
df['review']=df['review'].str.lower()
```

```python
# remove html tags
import re
def remove_html_tags(text):
  pattern=re.compile('<.*?>')
  return pattern.sub(r'',text)
remove_html_tags(df['review'][0])
```

```python
df['review']=df['review'].apply(remove_html_tags)
```

```python
# remove urls if any
def remove_urls(text):
  pattern=re.compile('https?://\S+|www\.\S+')
  return pattern.sub(r'',text)
remove_urls('hello i am tanys https://colab.research.google.com/drive/1T9HAtxKs9LS7xXHb0OmFNWbDOf1an6RG?usp=sharing')
```

```python
df['review']=df['review'].apply(remove_urls)
```

```python
# removing puntuations
import string
exclude=string.punctuation
def remove_punc(text):
  for char in exclude:
    text=text.replace(char,'')
  return text
remove_punc('HEllo! how are you? hope you are fine ** ')
```

```python
df['review']=df['review'].apply(remove_punc)
```

```python
def remove_chat_words(text):
  new_text=[]
  for word in text.split():
    if word.upper() in chat_words:
      new_text.append(chat_words[word.upper()])
    else:
      new_text.append(word)
  return " ".join(new_text)
print(remove_chat_words('THX without you this could have not been done'))
remove_chat_words('FYI meeting is now, call me ASAP')
```

```python
df['review']=df['review'].apply(remove_chat_words)
```

```python
# stop words removal
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
def remove_stopwords(text):
  new_text=[]
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  return " ".join(new_text)
remove_stopwords('hello! i am here')
```

```python
df['review']=df['review'].apply(remove_stopwords)
```

```python
# we can tokenize after stemming
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
stem_words('this is a big day. bcoz its your wedding')
```

```python
df['stem_reviews']=df['review'].apply(stem_words)
```

```python
# tokenization into words
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
word_tokenize('this is a book')
```

```python
def tokenize(text):
  return word_tokenize(text)
# word_tokenize(df['review'][0])
```

```python
df['words_review']=df['stem_reviews'].apply(tokenize)
```
